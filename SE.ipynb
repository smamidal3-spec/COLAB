```python
def temperature_modeling(a, b, c, time):
    temperature = a * time**2 + b * time + c
    return temperature


a_hardcoded, b_hardcoded, c_hardcoded = 0.1, 2, 10

print("Hard-coded Variables for Weather Modeling")
time_hardcoded = 5
print(
    "Temperature for hardcoded coefficients at time",
    time_hardcoded,
    "hours:",
    temperature_modeling(a_hardcoded, b_hardcoded, c_hardcoded, time_hardcoded)
)
print("\n")


print("Keyboard Input for Weather Modeling")
a_keyboard = float(input("Enter coefficient a: "))
b_keyboard = float(input("Enter coefficient b: "))
c_keyboard = float(input("Enter coefficient c: "))
time_keyboard = float(input("Enter time in hours: "))

print(
    "Temperature for keyboard input coefficients at time",
    time_keyboard,
    "hours:",
    temperature_modeling(a_keyboard, b_keyboard, c_keyboard, time_keyboard)
)
print("\n")


with open("weather_data.txt", "w") as file:
    file.write("0.2 1.5 8\n")
    file.write("4\n")


with open("weather_data.txt", "r") as file:
    lines = file.readlines()
    a_file, b_file, c_file = map(float, lines[0].split())
    time_file = float(lines[1])


print("Read from File for Weather Modeling")
print(
    "Temperature for file input coefficients at time",
    time_file,
    "hours:",
    temperature_modeling(a_file, b_file, c_file, time_file)
)
print("\n")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Quadratic model function
def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

# Generate weather data with optional noise
def generate_weather_data(start, end, num_points, a, b, c, noise_factor=10):
    x = np.linspace(start, end, num_points)
    y = quadratic_model(x, a, b, c) + np.random.normal(0, noise_factor, num_points)
    return x, y

# Plot weather data
def plot_weather_data(x, y, title):
    plt.scatter(x, y, label='Weather Data')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.show()

# Example parameters for the quadratic model
a = 0.5
b = -3
c = 20

# Generate weather data
start_time = 0
end_time = 10
num_points = 100

x_data, y_data = generate_weather_data(start_time, end_time, num_points, a, b, c)

# Plot the weather data (labeled "Waterfall Model" as per your code)
plot_weather_data(x_data, y_data, 'Waterfall Model: Weather Data')

# Display the final temperature value
final_temperature = y_data[-1]
print(f"\nFinal Temperature for Waterfall Model (Example Parameters): {final_temperature}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Quadratic temperature model
def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

# Generate synthetic weather data
def generate_weather_data(start, end, num_points, a, b, c, noise_factor=10):
    x = np.linspace(start, end, num_points)
    y = quadratic_model(x, a, b, c) + np.random.normal(0, noise_factor, num_points)
    return x, y

# Plot weather data
def plot_weather_data(x, y, title):
    plt.scatter(x, y, label='Weather Data')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.title(title)
    plt.legend()
    plt.show()

# Initial parameters for the quadratic model
a_initial = 0.5
b_initial = -5
c_initial = 20

# Generate initial weather data
start_time = 0
end_time = 10
num_points = 100

x_data, y_initial = generate_weather_data(
    start_time, end_time, num_points,
    a_initial, b_initial, c_initial
)

# Initial plot
plot_weather_data(
    x_data,
    y_initial,
    'Iterative Model: Initial Weather Data'
)

# Iterative development
num_iterations = 3  # Adjust as needed

for iteration in range(1, num_iterations + 1):
    print(f"\nIteration {iteration}: Gather feedback and refine the model.")

    a_iterative = float(input("Enter new value for 'a': "))
    b_iterative = float(input("Enter new value for 'b': "))
    c_iterative = float(input("Enter new value for 'c': "))

    # Generate updated weather data
    y_iterative = quadratic_model(
        x_data,
        a_iterative,
        b_iterative,
        c_iterative
    )

    # Plot updated data
    plot_weather_data(
        x_data,
        y_iterative,
        f'Iterative Model: Iteration {iteration} Weather Data'
    )

    if iteration == num_iterations:
        print(
            f"\nTemperature values for Iteration {iteration}:\n{y_iterative}"
        )

# Final plot
plot_weather_data(
    x_data,
    y_iterative,
    'Iterative Model: Final Weather Data'
)
```
